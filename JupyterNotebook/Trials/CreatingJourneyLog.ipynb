# Creating Journey Log

#### Creating the engine to connect to the database

```python
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine('postgresql+psycopg2://postgres:00001234@localhost:5433/jetaDb')
df_routes = pd.read_sql_query("select * from main_routes;",engine)
```

```python
df_routes
```

#### Creating a new dataframe to hold all the routes with segment

```python
df_journey = pd.DataFrame(columns=['routeid','segmentno','segmentid'])
```

```python
routeids = []
segmentnos = []
segmentids = []
```

```python
for i, rows in df_routes.iterrows():
    for j in range(len(rows['stopids'])):
        if j < (len(rows['stopids'])-1):
            segment = str(rows['stopids'][j])+"_"+str(rows['stopids'][j+1])
            segmentids.append(segment)
            segmentnos.append(j+1)
            routeids.append(rows['routeid'])
        else:
            break
```
